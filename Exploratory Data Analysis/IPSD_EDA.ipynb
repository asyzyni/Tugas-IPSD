## filling missing values

### menggunakan modus

```python
df_new = df.copy() # buat salinan dari df agar tetap aman saat di modifikasi 


# Yang menggunakan Modus, karena tipe datanya object

## Gender
mode_gender = df['Gender'].mode()[0] # ambil nilai modus pada kolom gender
df_new['Gender'] = df['Gender'].fillna(mode_gender)

## Blood Type 
mode_blood = df['Blood Type'].mode()[0] 
df_new['Blood Type'] = df['Blood Type'].fillna(mode_blood)

## Smoking 
mode_smoke = df['Smoking'].mode()[0]
df_new['Smoking'] = df['Smoking'].fillna(mode_smoke)

## Cholesterol
mode_chol = df['Cholesterol'].mode()[0]
df_new['Cholesterol'] = df['Cholesterol'].fillna(mode_chol)

## diabetes 
mode_diabet = df['Diabetes'].mode()[0]
df_new['Diabetes']= df['Diabetes'].fillna(mode_diabet)
```

### menggunakan modus/median

#### cek distribusi dan outliers

```python
# mengecek distribusi dan outliers 

## outliers 
df_numerical = df_new[['Age', 'Height', 'Weight', 'BMI', 'Temperature', 'Heart Rate', 'Blood Pressure']]

# Flag untuk melacak apakah ada outliers
outliers_found = False

# Loop untuk memeriksa setiap kolom
for col in df_numerical:
    Q1 = df_new[col].quantile(0.25)  # Kuartil pertama
    Q3 = df_new[col].quantile(0.75)  # Kuartil ketiga
    IQR = Q3 - Q1  # Rentang Interkuartil
    
    # Tentukan batas bawah dan atas untuk outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Deteksi outliers
    outliers = df_new[(df_new[col] < lower_bound) | (df_new[col] > upper_bound)]
    
    if not outliers.empty:  # Jika ada outliers
        print(f"Outliers ditemukan di kolom {col}")
        outliers_found = True

# Jika tidak ada outliers di semua kolom
if not outliers_found:
    print("Tidak ada outliers yang ditemukan di kolom-kolom yang diperiksa.")



```

```python
not_outlier = df_new[['Age', 'Height', 'Weight', 'Heart Rate', 'Blood Pressure']]
df_outlier = df_new[['BMI', 'Temperature']]

## untuk not_outlier 
for col in not_outlier.select_dtypes(include=['float64', 'int64']).columns:
    mean_val = not_outlier[col].mean()  # Menghitung rata-rata (mean)
    
    # Mengisi missing values dengan mean ke df_new
    df_new[col] = df_new[col].fillna(mean_val)

## untuk outlier 
for column in df_outlier.select_dtypes(include=['float64', 'int64']).columns:
    median_val = df_outlier[column].median() # mencari median 
    df_new[column] = df_new[column].fillna(median_val) #mengisi missing values
```

## Transformasi Data

```python
df_new['Height'] = df_new['Height'].round(0).astype(float)
df_new['Weight'] = df_new['Weight'].round(0).astype(float)
df_new['Cholesterol'] = df_new['Cholesterol'].round(0).astype(float)
df_new['Heart Rate'] = df_new['Heart Rate'].round(0).astype(float)
```

## Visualisasi Data

### Boxplot

```python
#hapus Student ID terlebih dahulu
df_graphics = df_new[['Age', 'Height', 'Weight', 'BMI', 'Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']]

plt.figure(figsize=(15, 8))

# Membuat boxplot untuk setiap kolom numerik
for i, col in enumerate(df_graphics.select_dtypes(include=['float64', 'int64']).columns):
    plt.subplot(1, len(df_graphics.select_dtypes(include=['float64', 'int64']).columns), i + 1)
    plt.boxplot(df_graphics[col])  # Mengabaikan nilai NaN
    plt.title(col)
    plt.xlabel("Nilai")
    
plt.suptitle("Boxplot untuk Semua Kolom Numerik", fontsize=16)
plt.tight_layout()
plt.show()
```

 ### Korelasi beserta heatmap

```python
correlation = df_graphics.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot = True, annot_kws ={'size' : 10}, cmap = 'coolwarm', cbar=True, fmt='.2f', square=False)
plt.show()
```

## Unique Values ## 

```python
#nilai unik
unique_counts = {} #buat dictionary kosong untuk menaruh hasil iterasi

for column in df_new.columns: #looping untuk mengecek semua kolom pada df
    unique_count = df_new[column].nunique() #digunakan untuk mengecek nilai unik di setiap kolom df
    unique_counts[column] = unique_count 

persebaran_uniq_df = pd.DataFrame(list(unique_counts.items()), columns = ['Kolom', 'Nilai Unik']) #untuk mengubah dictionary unique_counts menjadi dataframe

print("persebaran nilai unik : ")
persebaran_uniq_df

```

```python
for_uniqval = df_new.drop(columns=['Student ID', 'Weight', 'Height', 'Temperature', 'BMI', 'Cholesterol', 'Blood Pressure'])

for col in for_uniqval.columns:
    plt.figure(figsize=(10, 6))
    unique_counts = for_uniqval[col].value_counts(dropna=False)  # Counting NaN values as well
    sns.barplot(x=unique_counts.index.astype(str), y=unique_counts.values)
    plt.xticks(rotation=45)
    plt.title(f'Unique Values and Counts in {col}')
    plt.ylabel('Count')
    plt.xlabel(col)
    plt.savefig(f'{col}_unique_values_plot.png', bbox_inches='tight')
    plt.show()
```
